### подключаем библиотеки

```python
import pandas as pd
import requests
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import plotly.express as px
from dotenv import load_dotenv
```

### получаем данные

```python
start_date = '2023-03-01'
end_date = '2023-09-01'

url_visits = f'https://data-charts-api.hexlet.app/visits?begin={start_date}&end={end_date}'
url_registrations = f'https://data-charts-api.hexlet.app/registrations?begin={start_date}&end={end_date}'

response_visits = requests.get(url_visits)
data_visits = response_visits.json()

response_registrations = requests.get(url_registrations)
data_registrations = response_registrations.json()
```

### изучаем данные

```python
df_visits = pd.DataFrame(data_visits)
df_visits.head(10)
```

```python
df_visits.tail(10)
```
